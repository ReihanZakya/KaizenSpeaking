```python
from transformers import WhisperForConditionalGeneration, WhisperTokenizer, WhisperProcessor

try:
    model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
    print("Model Whisper small berhasil dimuat.")

    tokenizer = WhisperTokenizer.from_pretrained("openai/whisper-small", language="Indonesian", task="transcribe")
    print("Tokenizer berhasil dimuat.")

    processor = WhisperProcessor.from_pretrained("openai/whisper-small", language="Indonesian", task="transcribe")
    print("Processor berhasil dimuat.")
except Exception as e:
    print("Gagal memuat model, tokenizer, atau processor:", str(e))
```

```python
from datasets import load_dataset
#Pake yg tadi akuu

try:
    common_voice = load_dataset("mozilla-foundation/common_voice_11_0", "id", split="train+validation", trust_remote_code=True)
    print("Dataset Common Voice bahasa Indonesia berhasil dimuat. Jumlah sampel:", len(common_voice))
except Exception as e:
    print("Gagal memuat dataset:", str(e))

```

```python
import torch

def preprocess_function(batch):
    try:
        sentences = batch["sentence"]
        audios = batch["audio"]

        inputs = tokenizer(sentences, return_tensors="pt", padding="max_length", truncation=True, max_length=256)

        max_length = 160000

        audio_arrays = []
        for audio in audios:
            audio_tensor = torch.tensor(audio["array"]).float()
            if audio_tensor.shape[0] > max_length:
                audio_tensor = audio_tensor[:max_length]
            else:
                padding = torch.zeros(max_length - audio_tensor.shape[0])
                audio_tensor = torch.cat((audio_tensor, padding))
            audio_arrays.append(audio_tensor)

        return {
            "input_ids": inputs["input_ids"],
            "attention_mask": inputs["attention_mask"],
            "audio": torch.stack(audio_arrays)
        }
    except Exception as e:
        print("Error saat preprocessing data:", str(e))
        return None  # Mengembalikan None jika terjadi error
```

```python
try:
    common_voice = common_voice.map(preprocess_function, batched=True)
    print("Preprocessing dataset selesai.")
except Exception as e:
    print("Gagal memproses dataset:", str(e))
```

```python
# Mengecek apakah ada elemen None dalam dataset setelah preprocessing
print("Memeriksa apakah ada elemen None dalam dataset...")

# Menghapus elemen None dari dataset
common_voice = common_voice.filter(lambda x: x["input_ids"] is not None and x["audio"] is not None)
print("Dataset sudah divalidasi dan elemen None dihapus.")
```

```python
import torch

def validate_data(batch):
    try:
        # Pastikan batch memiliki key "input_ids" dan "audio"
        if "input_ids" not in batch or "audio" not in batch:
            raise KeyError("Batch harus memiliki kunci 'input_ids' dan 'audio'.")

        # Konversi ke tensor jika belum berupa Tensor
        input_ids = torch.tensor(batch["input_ids"]) if not isinstance(batch["input_ids"], torch.Tensor) else batch["input_ids"]
        audio = torch.tensor(batch["audio"]) if not isinstance(batch["audio"], torch.Tensor) else batch["audio"]

        # Pastikan input_ids memiliki dimensi 2
        if input_ids.ndim == 1:
            input_ids = input_ids.unsqueeze(0)  # Tambahkan dimensi batch
        elif input_ids.ndim != 2:
            raise ValueError(f"Dimensi 'input_ids' tidak sesuai: {input_ids.ndim}, harus 2D.")

        # Pastikan audio memiliki dimensi 2
        if audio.ndim == 1:
            audio = audio.unsqueeze(0)  # Tambahkan dimensi batch
        elif audio.ndim != 2:
            raise ValueError(f"Dimensi 'audio' tidak sesuai: {audio.ndim}, harus 2D.")

        print("Data divalidasi dengan shape:", input_ids.shape, audio.shape)
        return True
    except Exception as e:
        print("Error saat validasi data:", str(e))
        return False

# Filter data valid dari common_voice
try:
    valid_data = common_voice.filter(validate_data)
    print("Data valid berhasil difilter.")
except AttributeError as e:
    print("Error:", str(e), "\nPastikan common_voice mendukung operasi .filter()")
```

**Ini nambahin F1 score sama Training**

```python
from evaluate import load
from datasets import Dataset

# Memuat metrik evaluasi
metric_f1 = load("f1")
metric_accuracy = load("accuracy")

# Fungsi evaluasi
def evaluate_model(eval_dataset, model, processor, device):
    predictions = []
    references = []

    for batch in eval_dataset:
        # Proses audio dengan pipeline
        inputs = processor.feature_extractor(batch["audio"]["array"], sampling_rate=16000, return_tensors="pt").to(device)
        with torch.no_grad():
            logits = model(**inputs).logits

        # Decode hasil prediksi
        pred_ids = torch.argmax(logits, dim=-1)
        pred_sentence = processor.tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
        predictions.extend(pred_sentence)

        # Tambahkan label asli
        references.append(batch["sentence"])

    # Evaluasi menggunakan metrik
    f1_score = metric_f1.compute(predictions=predictions, references=references, average="weighted")
    accuracy = metric_accuracy.compute(predictions=predictions, references=references)

    return f1_score, accuracy

# Pastikan split validasi
eval_split = common_voice.train_test_split(test_size=0.2)["test"]
f1, acc = evaluate_model(eval_split, model, processor, device)
print(f"F1 Score: {f1}")
print(f"Accuracy: {acc}")
```

```python
import requests

def transcribe_audio(audio_url):
    start_time = time.time()

    # Unduh file audio
    audio_file = "temp_audio.mp3"
    try:
        response = requests.get(audio_url, stream=True)
        with open(audio_file, "wb") as f:
            f.write(response.content)
        print("File audio berhasil diunduh.")
    except Exception as e:
        print("Gagal mengunduh audio:", str(e))
        return None

    # Transkripsi menggunakan pipeline
    try:
        transcription = pipe(audio_file, language="id")
        print("Transkripsi selesai.")
    except Exception as e:
        print("Gagal melakukan transkripsi:", str(e))
        transcription = None

    end_time = time.time()
    duration = end_time - start_time
    print(f"Durasi Transkripsi: {duration:.2f} detik")
    return transcription["text"] if transcription else None

# Test model dengan file audio
audio_url = "https://drive.google.com/uc?export=download&id=1gmMwJ8GaJBVERCoUwapi1O4_XoXOX6jt"
transcription = transcribe_audio(audio_url)
print("Hasil Transkripsi:", transcription)
```
